# Experimentation
We have utilized four distinct models in our analysis: Random Forest (rf), Decision Tree (dt), Logistic Regression (lr), and LightGBM. Each of these models represents a different approach to solving the problem or making predictions based on the data provided.

- **Random Forest (rf)**: This model constructs multiple decision trees during training and outputs the mode of the classes (classification) or the mean prediction (regression) of the individual trees.
  
- **Decision Tree (dt)**: This model is a flowchart-like structure where each internal node represents a "test" on an attribute, each branch represents the outcome of the test, and each leaf node represents a class label or a numerical value.

- **Logistic Regression (lr)**: This model is a statistical method used for binary classification. It estimates the probability that a given instance belongs to a particular class based on its features, using the logistic function.

- **LightGBM**: LightGBM is a gradient boosting framework that uses tree-based learning algorithms. It grows trees vertically, leaf-wise, instead of level-wise like many other popular tree-based models. This method tends to achieve faster training times and lower memory usage.

Each model has its own strengths and weaknesses, and the choice of model depends on factors such as the nature of the data, the problem at hand, and the desired outcome of the analysis. By employing multiple models, we can compare their performance and select the one that best suits our specific objectives.

```python
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc, roc_auc_score
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import joblib
import lightgbm as lgb
import numpy as np

# Assuming xv_train, xv_test, y_train, y_test are transformed training and testing data

# Define models
rf_model = RandomForestClassifier(n_estimators=55, random_state=42)
lr_model = LogisticRegression(
    random_state=42,  # Change the random state for reproducibility
    C=0.9,  # Regularization parameter (default is 1.0)
    solver='lbfgs',  # Optimization algorithm (default is 'lbfgs')
    max_iter=12,  # Maximum number of iterations (default is 100)
    penalty='l2'  # Regularization type (default is 'l2')
)
dt_model = DecisionTreeClassifier(random_state=42)
lgb_model = lgb.LGBMClassifier(random_state=41)

# Create pipelines for each model with SMOTE
rf_pipeline = ImbPipeline([
    ('sampling', SMOTE(random_state=42)),
    ('model', rf_model)
])

lr_pipeline = ImbPipeline([
    ('sampling', SMOTE(random_state=42)),
    ('model', lr_model)
])

dt_pipeline = ImbPipeline([
    ('sampling', SMOTE(random_state=42)),
    ('model', dt_model)
])

lgb_pipeline = ImbPipeline([
    ('sampling', SMOTE(random_state=42)),
    ('model', lgb_model)
])

# List of pipelines for ease of iteration
pipelines = [rf_pipeline, lr_pipeline, dt_pipeline, lgb_pipeline]

# Iterate over pipelines
for pipeline, model_name in zip(pipelines, ['Random Forest', 'Logistic Regression', 'Decision Tree', 'LightGBM']):
    print(f"\nTraining {model_name}...")

    # Train the model
    pipeline.fit(X_train, y_train)

    # Make predictions
    predictions = pipeline.predict(X_test)

    # Evaluate the model
    print(f"\nConfusion Matrix for {model_name}:")
    print(confusion_matrix(y_test, predictions))

    print(f"\nClassification Report for {model_name}:")
    print(classification_report(y_test, predictions))

    # Plot Confusion Matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, predictions), display_labels=np.unique(y_test))
    disp.plot(cmap=plt.cm.Blues, values_format='d')
    disp.ax_.set_title(f'Confusion Matrix - {model_name}')



plt.show()  # Show all plots together
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming rf_pipeline, lr_pipeline, dt_pipeline, and lgb_pipeline are your trained models

# Assuming X_test and y_test are your test data and labels

# Make predictions for each model
rf_y_score = rf_pipeline.predict_proba(X_test)[:, 1]
lr_y_score = lr_pipeline.predict_proba(X_test)[:, 1]
dt_y_score = dt_pipeline.predict_proba(X_test)[:, 1]
lgb_y_score = lgb_pipeline.predict_proba(X_test)[:, 1]

# Calculate ROC curve for each model
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_y_score)
lr_fpr, lr_tpr, _ = roc_curve(y_test, lr_y_score)
dt_fpr, dt_tpr, _ = roc_curve(y_test, dt_y_score)
lgb_fpr, lgb_tpr, _ = roc_curve(y_test, lgb_y_score)

# Calculate AUC for each model
rf_auc = auc(rf_fpr, rf_tpr)
lr_auc = auc(lr_fpr, lr_tpr)
dt_auc = auc(dt_fpr, dt_tpr)
lgb_auc = auc(lgb_fpr, lgb_tpr)

# Plot ROC curves
plt.figure(figsize=(8, 8))
plt.plot(rf_fpr, rf_tpr, label=f'Random Forest (AUC = {rf_auc:.2f})')
plt.plot(lr_fpr, lr_tpr, label=f'Logistic Regression (AUC = {lr_auc:.2f})')
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})')
plt.plot(lgb_fpr, lgb_tpr, label=f'LGBM (AUC = {lgb_auc:.2f})')

# Plot the diagonal line representing a random classifier
plt.plot([0, 1], [0, 1], 'k--', label='Random')

# Set plot labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
